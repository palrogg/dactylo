# Test API

Python pipeline should be able to add / patch custom texts to the Laravel API

```python
import httpx as h
import pandas as pd
```

```python
ENDPOINT = "http://127.0.0.1:8000"
HEADERS = {"Accept": "application/json"}

r = h.get(f"{ENDPOINT}/api/texts", headers=HEADERS)
r.json()
```

```python
pd.set_option("display.max_colwidth", 300)
pd.DataFrame(r.json()["trace"])
```

```python
data = {
    "title": "Title",
    "description": "Great great stuff",
    "author": "Yann-Arthus",
    "publicationYear": 1980,
    "content": "This is it.\nThis is it!\nThis…",
    "displayDate": "2025-07-18T14:09:47.066Z",
}
POST_HEADERS = {"Content-Type": "application/json"}
r = h.post(f"{ENDPOINT}/api/texts", json=data, headers=POST_HEADERS)
r
```
